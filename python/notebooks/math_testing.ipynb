```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def nmda(v):
    return ((v + 60) / 60) ** 2 / (1 + ((v + 60) / 60) ** 2)

# Plot NMDA function
v = np.linspace(-100, 100, 1000)
plt.plot(v, nmda(v))
plt.title("NMDA Receptor Voltage Dependence")
plt.xlabel("Membrane Potential (mV)")
plt.ylabel("NMDA Conductance Factor")
plt.grid()
plt.show()
```
