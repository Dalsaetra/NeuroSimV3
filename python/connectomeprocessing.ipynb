```python
df = pd.read_csv("normalized_weighted_data.csv", index_col=0)
df = df.drop(columns=["corticcortical"], errors="ignore")

# ------------------------------------------------------------------
# 2.  Define model types and layers  -------------------------------
# ------------------------------------------------------------------
MODEL_TYPES = ["nb1", "p23", "b", "nb", "ss4",
               "p4", "p5_p6", "TC", "TI", "TRN"]
TYPE_INDEX = {t: i for i, t in enumerate(MODEL_TYPES)}

LAYER_INDEX = dict(L1=0, L23=1, L4=2, L5=3, L6=4, Th=5)   # Th = Thalamus
N_LAYERS = 6
N_TYPES  = len(MODEL_TYPES)

# ------------------------------------------------------------------
# 3.  Helpers: map each raw label to (model-type, layer) ------------
# ------------------------------------------------------------------
def classify(label: str):
    """Return (model_type, layer_idx) or (None, None) if not used"""
    if label == "nb1":
        return "nb1", LAYER_INDEX['L1']

    if label.startswith("p23"):
        return "p23", LAYER_INDEX['L23']

    if label.startswith("b23"):
        return "b",   LAYER_INDEX['L23']
    if label.startswith("nb23"):
        return "nb",  LAYER_INDEX['L23']

    if label.startswith("ss4"):
        return "ss4", LAYER_INDEX['L4']

    if label.startswith("p4"):
        return "p4",  LAYER_INDEX['L4']
    if label == "b4":
        return "b",   LAYER_INDEX['L4']
    if label == "nb4":
        return "nb",  LAYER_INDEX['L4']

    if label.startswith("p5"):
        return "p5_p6", LAYER_INDEX['L5']
    if label == "b5":
        return "b",      LAYER_INDEX['L5']
    if label == "nb5":
        return "nb",     LAYER_INDEX['L5']

    if label.startswith("p6"):
        return "p5_p6", LAYER_INDEX['L6']
    if label == "b6":
        return "b",      LAYER_INDEX['L6']
    if label == "nb6":
        return "nb",     LAYER_INDEX['L6']

    if label in ("TCs", "TCn"):
        return "TC",  LAYER_INDEX['Th']
    if label in ("TIs", "TIn"):
        return "TI",  LAYER_INDEX['Th']
    if label == "TRN":
        return "TRN", LAYER_INDEX['Th']

    # anything else (incl. corticocortical) â€” ignore
    return None, None


# ------------------------------------------------------------------
# 4.  Build lookup tables: (layer, type) âžœ list of labels -----------
# ------------------------------------------------------------------
rows_of = {(lay, typ): [] for lay in range(N_LAYERS) for typ in MODEL_TYPES}
cols_of = {(lay, typ): [] for lay in range(N_LAYERS) for typ in MODEL_TYPES}

for r in df.index:
    typ, lay = classify(r)
    if typ is not None:
        rows_of[(lay, typ)].append(r)

for c in df.columns:
    typ, lay = classify(c)
    if typ is not None:
        cols_of[(lay, typ)].append(c)

# ------------------------------------------------------------------
# 5.  Allocate the big (fromL, toL, fromT, toT) array ---------------
# ------------------------------------------------------------------
C = np.zeros((N_LAYERS, N_LAYERS, N_TYPES, N_TYPES))

# ------------------------------------------------------------------
# 6.  Fill it with mean weights  -----------------------------------
# ------------------------------------------------------------------
for from_layer in range(N_LAYERS):
    for to_layer in range(N_LAYERS):
        for from_type in MODEL_TYPES:
            cols = cols_of[(from_layer, from_type)]
            if not cols:
                continue
            for to_type in MODEL_TYPES:
                rows = rows_of[(to_layer, to_type)]
                if not rows:
                    continue
                block = df.loc[rows, cols].to_numpy()
                if block.size:
                    C[from_layer,
                      to_layer,
                      TYPE_INDEX[from_type],
                      TYPE_INDEX[to_type]] = block.mean()
```
